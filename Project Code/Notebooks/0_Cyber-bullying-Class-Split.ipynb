**Reading and exploring the dataset**

```python
# Loading the dataset formsping.csv
data = pd.read_csv('/content/drive/MyDrive/Cyberbullying_Detection/formspring.csv')
data.head()
```

```python
# Getting the shape and the columns of the dataset
print(f'The shape of the dataset is {data.shape}')
print(f'The columns within the dataset are: \n{data.columns}')
```

```python
# After looking carefully to the dataset we found that the column post contains Quistions and answers that are
# Splitted into the columns ques and ans so we are going to drop them
# Also we are going to drop the columns userid and asker as they don't give valuble info

data.drop(columns=['userid','ques','ans','asker'],inplace=True)
```

**Filter the Bullying rows according to the severity**

```python
data.severity1.value_counts()
```

```python
# Get the data that has a sevirty of 0
temp = data[(data['severity1'] != '0') & (data['severity2'] != '0') & (data['severity3'] != '0')]
temp
```

```python
# Drop the rows that has a post value of nan
temp = temp.drop(list(temp[temp.post.isna()].index))
temp
```

```python
# Drop the extracted bullying records that have the severity values of NULL
temp1 = temp[(temp['severity1'].isna()) & (temp['severity2'].isna()) & (temp['severity3'].isna())]
temp = temp.drop(list(temp1.index))
temp
```

**Process the bully# columns**

```python
# Count the data that are nulls that are exists in the data as strings "within bully1 col"
data.bully1.value_counts()[:50]
```

```python
# Count the data that are nulls that are exists in the data as strings "within bully2 col"
data.bully2.value_counts()[:50]
```

```python
# Count the data that are nulls that are exists in the data as strings "within bully3 col"
data.bully3.value_counts()[:50]
```

```python
# Fill the replace the strings in the following list with np.nan
bully_miss_list = ["Na", "N/A.", "no", "N/a", "na", "#NAME?", "n/A", "No"]
data[['bully1', 'bully2', 'bully3']] = data[['bully1', 'bully2', 'bully3']].replace(bully_miss_list, np.nan)
```

**Filter the bullying rows according to the "bully#" columns**

```python
# Get the list of indices where each bully# column value is not null
l1 = list(data[data.bully1.notna()].index)
l2 = list(data[data.bully2.notna()].index)
l3 = list(data[data.bully3.notna()].index)

# Combine the indices together without indices redunduncy
indices = list(set(l1+l2+l3))
bully_data = data.iloc[indices,:]
```

**Get the Bully Rows from the severity filtered data and the bully filtered data**

```python
# Get the indices of the severity# filtered data
sev_indices =list( temp.index)

# Get the indices of the bully# filtered data
bully_indices = list(bully_data.index)

# Combine the both indices
indices_for_bully = sorted(list(set(sev_indices + bully_indices)))

# Get the final dataset
bully_data = data.iloc[indices_for_bully,:]
```

```python
bully_data
```

```python
# Get the Not bully Data
not_bully_data = data.drop(bully_data.index).reset_index(drop=True)
not_bully_data
```

```python
# Drop the bully_data where the post is = "Yes" or "No"
bully_data = bully_data.drop(bully_data[bully_data['post'].isin(['Yes', 'No'])].index).reset_index(drop=True)
```

```python
# Save the datasets as a csv files
bully_data.to_csv('bully_data.csv', index=False)
not_bully_data.to_csv('not_bully_data.csv', index=False)
```
